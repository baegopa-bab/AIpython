```python
# 다음 데이터를 이용하여 1년 동안의 월별 평균 강수량을 꺾은선 그래프로 그려보세요.
months =  ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]
# 그래프 제목과 x, y축 레이블을 추가하세요.
plt.figure(figsize=(5, 3))
plt.plot(months, rainfall, label='busan')
plt.title('Monthly rainfall')
plt.plot(months, rainfall, 'b.:')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')

plt.ylim(0, 1.2*max(rainfall))

# 데이터 포인터로 값 표시
for i, value in enumerate(rainfall):
  plt.text(i,value+10, str(value), ha='center', va='bottom')


plt.legend()
plt.show()
```

```python
months =  ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
rainfall = [30, 40, 60, 80, 100, 150, 300, 280, 120, 70, 50, 35]

plt.figure(figsize=(5, 3))
plt.plot(months, rainfall, label='busan')
for i in range(len(rainfall)):
  plt.annotate(rainfall[i], xy=(months[i],rainfall[i]), xytext=(5,-10), textcoords='offset points')
plt.title('Monthly rainfall')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')
plt.legend()
plt.show()
```

```python
print(np.zeros((2, 3)))      # 2x3 크기의 0 배열
print(np.ones(5))            # 길이 5의 1 배열
print(np.arange(0, 10, 2))   # 0부터 10까지 2씩 증가
print(np.linspace(0, 1, 5))  # 0~1 사이를 5개로 나눈 배열
print(np.random.rand(3))     # 0~1 사이의 난수 3개
```

```python
arr = np.array([[1, 2, 3], [4, 5, 6]])

print("Shape:", arr.shape)    # (2, 3) 2행3열
print("Data Type:", arr.dtype)  # int64
print("Size:", arr.size)      # 6
print("Dimension:", arr.ndim) # 2
```

```python
#배열 연산
a = np.array([1, 2, 3])
b = np.array([10, 20, 30])

print(a + b)         # [11 22 33]
print(a * b)         # [10 40 90]
print(a ** 2)        # 제곱 [1 4 9]
print(np.sqrt(a))    # 제곱근
```

```python
#집계
arr = np.array([1, 2, 3, 4, 5])

print("Sum:", np.sum(arr))
print("Mean:", np.mean(arr))
print("Standard Deviation:", np.std(arr))
print("Min/Max:", np.min(arr), np.max(arr))
```

```python
#조건부 선택
arr = np.array([10, 20, 30, 40, 50])

print(arr[arr > 30])  # 30보다 큰 요소만
```

```python
#문제1: 다음 조건에 맞는 배열을 생성하고 출력
#1차원 배열, 5개의 요소, 0으로 초기화된 배열
arr=np.zeros(5)
print(arr)
```

```python
#문제2: 1부터 10까지의 숫자 중 짝수만 포함하는 배열 생성
arr = np.arange(2, 11, 2)
print(arr)
```

```python
# [2, 4, 6] 배열과 [10, 20, 30] 배열을 각각 생성하고 곱셈 연산을 수행한 결과를 출력

a = np.array([2, 4, 6])
b = np.array([10, 20, 30])

result = a * b
print(result)
```

```python
# 위 배열에서 20보다 큰 요소만 필터링해서 출력
filtered = result[result > 20]
print(filtered)
```
